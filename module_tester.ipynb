## Testing Trainer Modules

```python
import torch
from torch.utils.data import TensorDataset, DataLoader

from modules.architectures.Architecture import Architecture
from modules.helper.Trainer import Trainer

# Model
model = Architecture()
model.add(
    torch.nn.Linear(10, 1, bias=True)
).to("cuda")

# Dummy dataset
x_train = torch.randn(100, 10)   # 100 samples, 10 features
y_train = torch.randn(100, 1)    # 100 targets

x_test = torch.randn(100, 10)
y_test = torch.randn(100, 1)

# Dataset + DataLoader
train_dataset = TensorDataset(x_train, y_train)
test_dataset = TensorDataset(x_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=16)

# Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=1e-3)

# Trainer
trainer = Trainer(
    model=model,
    train_loader=train_loader,
    val_loader=test_loader,
    optimizer=optimizer,
    criteration=torch.nn.MSELoss(),
    print_every=1,
    device="cuda"
)

trainer.train(1)
```
