## Import Necessary Modules

```python
from pathlib import Path
import copy
import os
import shutil
import uuid
import random
```

```python
# setting a seed
random.seed(41)
```

## Merge Dataset

### Create Directories

```python
base_path = Path("./datasets/final_dataset")

folders  = [
    "क", "ख", "ग", "घ", "ङ",
    "च", "छ", "ज", "झ", "ञ",
    "ट", "ठ", "ड", "ढ", "ण",
    "त", "थ", "द", "ध", "न",
    "प", "फ", "ब", "भ", "म",
    "य", "र", "ल", "व", "श",
    "ष", "स", "ह", "क्ष", "त्र", "ज्ञ"
]

for folder in folders:
    (base_path / "train"/ folder).mkdir(exist_ok=True, parents=True)
    (base_path / "test"/ folder).mkdir(exist_ok=True, parents=True)
    (base_path / "val"/ folder).mkdir(exist_ok=True, parents=True)

print("Folders Created Successfully!")
```

### Copy Files Class Wise to merged

```python
for i in range(36):
    # input directories
    plain_bg_class = Path(f"./datasets/NSLv2/data/Plain Background/{i}")
    random_bg_class = Path(f"./datasets/NSLv2/data/Random Background/{i}")
    # all files
    all_files = []
    
    for item in plain_bg_class.iterdir():
        all_files.append(item)
    
    for item in random_bg_class.iterdir():
        all_files.append(item)
        
    random.shuffle(all_files)
    
    train_len = int(len(all_files) * 0.65)
    val_len = int(len(all_files) * 0.20)
    
    splits = {
    "train" : all_files[:train_len],
    "val" : all_files[train_len:train_len+val_len],
    "test": all_files[train_len+val_len :]
    }
    
    for split in splits:
        out_dir = Path(f"./datasets/final_dataset/{split}/{folders[i]}")
        print(out_dir)
        files = splits[split]
        for item in files:
            file_name = f"{uuid.uuid4()}-{uuid.uuid4()}-{folders[i]}.png"
            shutil.copy2(item, out_dir/file_name)
```
